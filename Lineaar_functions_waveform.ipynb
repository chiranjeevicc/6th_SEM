```python
# to plot a sine wave  sin(2pift)
t=np.linspace(0,1,100)    # 100 values in the interval 0 to 1
f=1                       # frequency of the sine wave
x_t=np.sin(2*np.pi*f*t)

plt.plot(t,x_t)
plt.grid()
plt.xlabel('time (t)')
plt.ylabel('sine function')
plt.title('plot of function')
```

```python
#Example I:To plot sum of sinusoid

t = np.linspace(0,2,60)
f1 = 1
f2 = 1
x1 = np.cos(2 * np.pi * f1 * t)
x2=np.sin(2 * np.pi * f2 * t)
x1_plus_x2 = x1 + x2

plt.figure(figsize=(15,10))
plt.plot(t,x1,label = 'x1(t)')
plt.plot(t,x2,label = 'x2(t)')
#plt.plot(t,x1_plus_x2, label = 'x1(t) + x2(t)',lw=5)
plt.plot(t,x1_plus_x2,label = 'x1(t) + x2(t)',lw=5)
plt.title('Sum of sinusoids')
plt.legend()
plt.grid
```

```python
#Example 2 sinusoids
t=np.linspace(0,2,60)
f1=1
f2=2
x1=np.cos(2*np.pi*f1*t)
x2=np.sin(2*np.pi*f2*t)
x1_plus_x2 = x1+x2

#plt.figure(figsize = (10,15))
plt.plot(t,x1, label = 'x1(t)')
plt.plot(t,x2, label = 'x2(t)')
plt.plot(t,x1_plus_x2, label = 'x1(t)+x2(t)',lw=5)
plt.title('Diff Frequency Sinusoids')
plt.grid()
plt.legend()
```

```python
# to plot a sinewave and its square 
t = np.linspace(0,1,60)
f1 = 1
x1 = np.sin(2 * np.pi * f1 * t)
x2 = x1*x1

plt.figure(figsize=(12,8))
plt.plot(t,x1, label='sinusoid')
plt.plot(t,x2, label = 'squared sinusoid',lw=5)

plt.legend()
plt.grid()
```

```python
# to plot a sinewave and its square 
t = np.linspace(0,1,60)
f1 = 1
x1 = np.cos(2 * np.pi * f1 * t)
x2 = x1*x1

plt.figure(figsize=(12,8))
plt.plot(t,x1, label='sinusoid')
plt.plot(t,x2, label = 'squared sinusoid',lw=5)

plt.legend()
plt.grid()
```

```python
# Example I: To plot the product of sinusoids
t = np.linspace(0,2,60)
f1 = 1
f2 = 1
x1 = np.cos(2 * np.pi * f1 * t)
x2 = np.sin(2 * np.pi * f2 * t)
x1_into_x2 = x1 * x2

plt.figure(figsize=(15,10))
plt.plot(t,x1, label='x1(t)')
plt.plot(t,x2, label = 'x2(t)')
plt.plot(t,x1_into_x2, label = 'x1(t) * x2(t)',lw=5)
plt.title('Product of two sinusoids')
plt.legend()
plt.grid()
```

```python
# Example II: To plot the product of sinusoids
t = np.linspace(0,1,300)
f1 = 1
f2 = 10
x1 = np.sin(2 * np.pi * f1 * t)
x2 = 0.95*np.sin(2 * np.pi * f2 * t)
x1_into_x2 = x1 * x2

plt.figure(figsize=(15,8))
plt.plot(t,x1, label='x1(t)')
plt.plot(t,x2, label = 'x2(t)')
plt.plot(t,x1_into_x2, label = 'x1(t) * x2(t)',lw=5)
plt.title('Product of two sinusoids')
plt.legend()
plt.grid()
```

```python
# to plot the sync function  sin(x)/(x)
# NOTE: here carefuly select t, to avoid division by zero
# if not possible,then there is a need to have an if statement to incorporate the change

t = np.linspace(-2,2,60)
x = 2 * np.pi * t
x1 = np.sin(x)

synx_x = x1/x

plt.figure(figsize=(15,8))
plt.plot(t,synx_x, label='x1(t)')
plt.title('Sync Function')
plt.legend()
plt.grid()
```

```python
# to plot a sinewave and its cube 
t = np.linspace(0,1,60)
f1 = 1
x1 = np.cos(2 * np.pi * f1 * t)
x2 = np.power(x1,3)

plt.figure(figsize=(12,8))
plt.plot(t,x1, label='sinusoid')
plt.plot(t,x2, label = 'cubed sinusoid',lw=5)

plt.legend()
plt.grid()
```

```python
# To plot an Exponentially decaying sinusoid

t = np.linspace(0,5,100)
x1 = np.sin(2 * np.pi *  t)
x2 = np.exp(-0.5* t)
x1_into_x2 = x1 * x2

plt.figure(figsize=(15,10))
plt.plot(t,x1, label='x1(t)')
plt.plot(t,x2, label = 'x2(t)')
plt.plot(t,x1_into_x2, label = 'x1(t) * x2(t)',lw=3)
plt.title('Exponentially decaying sinusoid')
plt.legend()
plt.grid()
```
